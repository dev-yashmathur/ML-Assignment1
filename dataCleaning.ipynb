```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("bpp_training_data.csv")
df
```

```python
col_list = df.columns
for col in col_list:
    na_values = df[col].isna().sum()
    print(col + " : " +  str(na_values), end="\n")
```

```python
for col in col_list:
    cnt = len(df[col].unique())
    print(col + " : " + str(cnt))
```

```python
y = df["SalePrice"]
df.drop('SalePrice', inplace=True, axis=1)
col_list2 = df.columns
plot = []
for i in range(55):
    plot.append(plt.figure(i))
    plt.scatter(df[col_list2[i]], y)
    plt.xlabel(col_list2[i])
    plt.ylabel("SalePrice")

plt.show()
```
